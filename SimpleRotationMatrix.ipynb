```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y = []

a = 6371 + 700

for v in range(0,360):
  x.append(a*np.cos( np.radians(v) ))
  y.append(a*np.sin(np.radians(v)))

plt.plot(x,y)
plt.axis('equal')  
```

```python
v = np.arange(0,360,1)

x = a*np.cos(np.radians(v))
y = a*np.sin(np.radians(v))
z = np.zeros_like(x)

I = 30
R1 = np.array(
    [
     [1 , 0, 0],
     [ 0 , np.cos(np.radians(-I)) , np.sin(np.radians(-I)) ],
     [ 0 , -np.sin(np.radians(-I)) , np.cos(np.radians(-I)) ]
    ]
)

C = np.vstack ( (x, y, z) )

P = np.dot(R1, C)


ax = plt.figure().add_subplot(projection='3d')
ax.plot(P[0,:], P[1,:], zs=P[2,:], zdir='z', label='curve in (x, y)')
```
